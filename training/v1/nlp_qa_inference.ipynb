```python
from transformers import pipeline

nlp = pipeline("question-answering", model="JohnOfGod33/distilbert-base-cased-squad-qa")
```

```python
questions = [{
    "context" :  """
Koffi is a backend developer working on a project called AgroPredict. The application helps farmers predict weather conditions. It was launched in 2024. Koffi works with Ama, a data scientist, and Yao, a backend developer. The project currently has 200 users.
""",
    "questions" : [
    "What is the name of the project?",
    "What is Koffi's job?",
    "When was the application launched?",
    "Who is Ama?",
    "How many users does the project have?"
    "who are joe biden ?"
]
}, {
    "context" : """Koffi est un développeur qui a créé une application appelée AgroPredict pour aider les agriculteurs à prévoir la météo. Il travaille avec Ama, une data scientist, et Yao, un développeur backend.

Au début, les données étaient difficiles à utiliser à cause de capteurs défectueux. Après amélioration, la précision des prévisions est passée de 60 % à 85 %.

L’application a été lancée en 2024 et a attiré 300 utilisateurs en deux mois. Une entreprise a proposé de racheter le projet, mais Koffi a refusé pour garder le contrôle et rester fidèle à son objectif d’aider les agriculteurs.""",
    "questions" : [
        "Comment s’appelle l’application ?",
        "Qui est Koffi ?",
        "Quand a été lancée l’application ?",
        "Quel problème affectait les données ?",
        "Quelle était la précision initiale des prévisions ?",
        "Combien d’utilisateurs ont rejoint l’application ?",
        "En quelle année l’application a-t-elle été lancée ?",
        "Pourquoi Koffi refuse-t-il de vendre son projet ?"
    ]
}]

print("--- DAILY LIFE INFERENCE ---")
threshold = 0.05 # Seuil de

for q in questions:
    print("="*30)
    context = q['context']
    print(f"context: {context}")

    questions = q['questions']
    for question in questions:
        res = nlp(question=question, context=context)
        if float(res['score']) < threshold:
            answer = "Je ne trouve pas de reponse"
        else:
            answer = res['answer']

        print(f"Q: {question}")
        print(f"R: {answer} (Score: {res['score']:.4f})")
        print("-" * 10)
```
